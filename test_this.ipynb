```python
import gymnasium as gym
from carl.envs import CARLCartPole
```

```python
env = CARLCartPole(contexts={
    "0": {"masspole": 0.1, "length": 0.5}}, render_mode="human")
env.metadata["render_mode"]
print(env.metadata)
obs = env.reset()
print(obs)
```

```python
for i in range(1000):
    action = env.action_space.sample()
    obs, reward, done, info, _ = env.step(action)
    env.render()
    print(f"Step: {i}, Obs: {obs}, Reward: {reward}, Done: {done}")
    if done:
        obs = env.reset()
        print("Environment reset")
```
